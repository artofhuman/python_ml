f(0)  = w0 * 0  + w1 * 0  + w2 * 0 ** 2
f(1)  = w0 * 1  + w1 * 1  + w2 * 1 ** 2
f(15) = w0 * 15 + w1 * 15 + w2 * 15 ** 2



```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)


def show_main_graph():
    x = np.arange(1, 16.);
    y = f(x)

    plt.plot(x, y)
    plt.grid(True) #Сетка
    plt.xlabel(r'$x$') #Метка по оси x в формате TeX
    plt.ylabel(r'$f(x)$') #Метка по оси y в формате TeX

    plt.axis([0, 16, 0, 3.5])
    
show_main_graph()
```

```python
def build_system(x0):
    A = []
    v = []
    for _x in x0:
        v.append(f(_x))
        A.append([_x**i for i in range(len(x0))])
        
    return A, v

def f1(x, w):
    i = 0
    res = 0
    for we in w:
        res += we * (x ** i)
        i += 1
    return res

A, v = build_system([1., 15.])
w = np.linalg.solve(A, v)

A2, v2 = build_system([1., 8., 15.])
w2 = np.linalg.solve(A2, v2)

A3, v3 = build_system([1., 4., 10., 15.])
w3 = np.linalg.solve(A3, v3)

print('A1', A, 'v1', v, 'w1', w)
print('A2', A2, 'v2', v2, 'w2', w2)
print('A3', A3, 'v3', v3, 'w3', w3)

x = np.arange(1., 16., 0.1);
y = f(x)

plt.plot(x, y)
plt.grid(True)
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')


plt.plot(x, f1(x, w))
plt.plot(x, f1(x, w2))
plt.plot(x, f1(x, w3))


plt.show()

print('Solution', w3)
```
