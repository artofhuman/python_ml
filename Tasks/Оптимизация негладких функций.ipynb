```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize



def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

def print_func_on_arange(x, func=f):
    y = []
    for i in x:
      y.append(func(i))
    
    plt.plot(x, y)
    plt.show()


print_func_on_arange(np.arange(1., 31., 1))
```

```python
res = optimize.minimize(f, 30., method='BFGS')
print(res.x)

print_func_on_arange(np.arange(1., 30., 0.1))
```

```python
res_2 = optimize.minimize(f, 2., method='BFGS')
res_30 = optimize.minimize(f, 30., method='BFGS')
print(
    round(f(res_2.x[0]), 2),
    round(f(res_30.x[0]), 2)
)

```

```python
# Глобальная оптимизация

bounds = (1, 30)
res = optimize.differential_evolution(f, [bounds])

print(round(f(res.x), 2))
```

```python
# Минимизация негладкой функции

def h(x):
    return int(f(x))

print_func_on_arange(np.arange(1., 32., 0.5), h)
```

```python
res = optimize.minimize(h, 30.0, method='BFGS')
print(h(res.x))

res = optimize.differential_evolution(h, [bounds])
print(h(res.x))
```
